# Minimax GSP: Python Translation Test

This notebook exercises the translated functions on a small random GSP Ising model.

```python
import numpy as np
from minimax_gsp import random_gsp_ising, exact_ising, find_gsp_update

np.random.seed(0)
n = 5
J_true, h_true = random_gsp_ising(n)
mean, corr, _, _ = exact_ising(J_true, h_true, kT=1.0)
h_hat, J_hat, ent, ents = find_gsp_update(mean, corr)

print('J difference (fro):', np.linalg.norm(J_hat - J_true))
print('h difference (l2):', np.linalg.norm(h_hat - h_true))
print('Final entropy estimate:', ent)
```
